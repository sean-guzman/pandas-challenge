In this analysis, data is taken from a school district where a series of calculations are performed to determine its success as a district, as well as by school, type of school, school spending.  The data is broken down even further by grade to uncover any trends as students progress through high school.  

## Key Takeaways and Conclusions
- As a district, it averaged a 78.99 for Math and 81.88 for Reading. While the majority passed each subject respectively (Math - 74.98%, Reading 85.81%), only 65% of students passed both.
- Scores for both Math and Reading appear to stay relatively the same throughout high school.
- Spending more per student does not necessarily yield higher Overall Passing Percentages as shown in the spending summary dataframe but given the data, schools with smaller student populations have a higher Overall Passing Percentage despite having smaller budgets as shown in the school summary dataframe.
- Further validating the previous point, the 5 highest-performing schools of passing overall are charter schools (where budget and size are both smaller), whereas the lowest-performing are district schools.  Charter schools excel in all categories [ Average Math Score, Average Reading Score, Passing Math Percentage, Passing Reading Percentage, Overall Passing Percentage ] over district schools as shown in the type_summary dataframe.
- Scores and percentages in all categories [ Average Math Score, Average Reading Score, Passing Math Percentage, Passing Reading Percentage, Overall Passing Percentage ] improve signifcantly depending on size/number of students as shown in the size_summary dataframe. Higher marks in these categories can be seen Medium size schools (1000-2000 students) versus Large size schools (over 2000 students).  However, little improvement can be seen, if any, when comparing Medium size schools to Small size schools (less than 1000 students).

### District Summary

```python
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Total number of unique schools
school_count = school_data['school_name'].count()
school_count
```

```python
# Total students
student_count = student_data['student_name'].count()
student_count
```

```python
# Total budget
total_budget = school_data['budget'].sum()
total_budget
```

```python
# Average math score
average_math_score = student_data['math_score'].mean()
average_math_score
```

```python
# Average reading score
average_reading_score = student_data['reading_score'].mean()
average_reading_score
```

```python
# Percentage passing math (the percentage of students who passed math)
passing_math_count = student_data[(student_data["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count/student_count * 100
passing_math_percentage
```

```python
# Percentage passing reading (the percentage of students who passed reading)
passing_reading_count = student_data[(student_data["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count/student_count * 100
passing_reading_percentage
```

```python
# Overall passing (the percentage of students who passed math AND reading)
overall_passing_count = student_data[(student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70)].count()["student_name"]
overall_passing_percentage = overall_passing_count/student_count * 100
overall_passing_percentage
```

```python
# District Summary displaying all found metrics in one dataframe

district_summary = pd.DataFrame({
    "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget" : [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "Passing Math Percentage": [passing_math_percentage],
    "Passing Reading Percentage": [passing_reading_percentage],
    "Overall Passing Perecntage": [overall_passing_percentage]
})

# Format to include , for thousands/millions, $ for monetary amounts, % for percentages, and round to two decimal places
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.2f}".format)
district_summary["Passing Math Percentage"] = district_summary["Passing Math Percentage"].map("{:,.2f}%".format)
district_summary["Passing Reading Percentage"] = district_summary["Passing Reading Percentage"].map("{:,.2f}%".format)
district_summary["Overall Passing Perecntage"] = district_summary["Overall Passing Perecntage"].map("{:,.2f}%".format)


district_summary
```

### Math Scores by Grade

```python
# Perform the necessary calculations to average math score for students of each grade level (9th, 10th, 11th, 12th) at each school.
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by "school_name" and take the mean of each.
ninth_graders_scores = ninth_graders.groupby(["school_name"]).mean()["math_score"]
tenth_graders_scores = tenth_graders.groupby(["school_name"]).mean()["math_score"]
eleventh_graders_scores = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
twelfth_graders_scores = twelfth_graders.groupby(["school_name"]).mean()["math_score"]

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame(columns=["9th", "10th", "11th", "12th"])
math_scores_by_grade["9th"] = ninth_graders_scores
math_scores_by_grade["10th"] = tenth_graders_scores
math_scores_by_grade["11th"] = eleventh_graders_scores
math_scores_by_grade["12th"] = twelfth_graders_scores

# Format averages to two decimal places
math_scores_by_grade["9th"] = math_scores_by_grade["9th"].map("{:,.2f}".format)
math_scores_by_grade["10th"] = math_scores_by_grade["10th"].map("{:,.2f}".format)
math_scores_by_grade["11th"] = math_scores_by_grade["11th"].map("{:,.2f}".format)
math_scores_by_grade["12th"] = math_scores_by_grade["12th"].map("{:,.2f}".format)

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

###  Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].str.replace("$","")

school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].astype("float")
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], spending_bins, labels=labels, include_lowest=True)
```

```python
# Remove $ and % from columns and re-assign type from string to float in order to calculate average
school_spending_df["Average Math Score"] = school_spending_df["Average Math Score"].astype("float")
school_spending_df["Average Reading Score"] = school_spending_df["Average Reading Score"].astype("float")

school_spending_df["Passing Math Percentage"] = school_spending_df["Passing Math Percentage"].str.replace("%","")
school_spending_df["Passing Math Percentage"] = school_spending_df["Passing Math Percentage"].astype("float")
school_spending_df["Passing Reading Percentage"] = school_spending_df["Passing Reading Percentage"].str.replace("%","")
school_spending_df["Passing Reading Percentage"] = school_spending_df["Passing Reading Percentage"].astype("float")
school_spending_df["Overall Passing Percentage"] = school_spending_df["Overall Passing Percentage"].str.replace("%","")
school_spending_df["Overall Passing Percentage"] = school_spending_df["Overall Passing Percentage"].astype("float")

school_spending_df
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Passing Math Percentage"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Passing Reading Percentage"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Overall Passing Percentage"]

```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "Passing Math Percentage": spending_passing_math,
    "Passing Reading Percentage": spending_passing_reading,
    "Overall Passing Percentage": overall_passing_spending
})

# Re-establish formatting adding % to percentages and rounding to two decimal places
spending_summary["Average Math Score"] = spending_summary["Average Math Score"].map("{:,.2f}".format)
spending_summary["Average Reading Score"] = spending_summary["Average Reading Score"].map("{:,.2f}".format)
spending_summary["Passing Math Percentage"] = spending_summary["Passing Math Percentage"].map("{:,.2f}%".format)
spending_summary["Passing Reading Percentage"] = spending_summary["Passing Reading Percentage"].map("{:,.2f}%".format)
spending_summary["Overall Passing Percentage"] = spending_summary["Overall Passing Percentage"].map("{:,.2f}%".format)


# Display results
spending_summary
```

### Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Create a copy of the school summary
school_size_df = per_school_summary.copy()
```

```python
# Remove ',' from "Total Students" and convert to int in order to categorize.
#Categorize the spending based on the bins.
school_size_df["Total Students"] = school_size_df["Total Students"].str.replace(",","")
school_size_df["Total Students"] = school_size_df["Total Students"].astype("int")
school_size_df["School Size"] = pd.cut(school_size_df["Total Students"], size_bins, labels=labels, include_lowest=True)
school_size_df
```

```python
school_size_df["Average Math Score"] = school_size_df["Average Math Score"].astype("float")
school_size_df["Average Reading Score"] = school_size_df["Average Reading Score"].astype("float")

school_size_df["Passing Math Percentage"] = school_size_df["Passing Math Percentage"].str.replace("%","")
school_size_df["Passing Math Percentage"] = school_size_df["Passing Math Percentage"].astype("float")
school_size_df["Passing Reading Percentage"] = school_size_df["Passing Reading Percentage"].str.replace("%","")
school_size_df["Passing Reading Percentage"] = school_size_df["Passing Reading Percentage"].astype("float")
school_size_df["Overall Passing Percentage"] = school_size_df["Overall Passing Percentage"].str.replace("%","")
school_size_df["Overall Passing Percentage"] = school_size_df["Overall Passing Percentage"].astype("float")


# Calculate averages for the desired columns. 
size_math_scores = school_size_df.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = school_size_df.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = school_size_df.groupby(["School Size"]).mean()["Passing Math Percentage"]
size_passing_reading = school_size_df.groupby(["School Size"]).mean()["Passing Reading Percentage"]
size_overall_passing = school_size_df.groupby(["School Size"]).mean()["Overall Passing Percentage"]
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({
    "Average Math Score": size_math_scores,
    "Average Reading Score": size_reading_scores,
    "Passing Math Percentage": size_passing_math,
    "Passing Reading Percentage": size_passing_reading,
    "Overall Passing Percentage": size_overall_passing
})

# Re-establish formatting adding % to percentages and rounding to two decimal places
size_summary["Average Math Score"] = size_summary["Average Math Score"].map("{:,.2f}".format)
size_summary["Average Reading Score"] = size_summary["Average Reading Score"].map("{:,.2f}".format)
size_summary["Passing Math Percentage"] = size_summary["Passing Math Percentage"].map("{:,.2f}%".format)
size_summary["Passing Reading Percentage"] = size_summary["Passing Reading Percentage"].map("{:,.2f}%".format)
size_summary["Overall Passing Percentage"] = size_summary["Overall Passing Percentage"].map("{:,.2f}%".format)


# Display results
size_summary
```
